```python
import pandas as pd
```

```python
windows_laptop_data = pd.concat(
    map(pd.read_csv, ['raw_data/windows_laptop_0417.csv', 'raw_data/windows_laptop_1820.csv', 'raw_data/windows_laptop_2122.csv']), ignore_index=True)
windows_laptop_data['OS'] = 'Windows'
windows_laptop_data.sort_values(by = 'Year')
```

```python
MacOS_laptop_data = pd.read_csv('raw_data/MacOS_laptop.csv')
MacOS_laptop_data['OS'] = 'MacOS'
MacOS_laptop_data.sort_values(by='Year')
```

```python
Linux_laptop_data = pd.read_csv('raw_data/Linux_laptop.csv')
Linux_laptop_data['OS'] = 'Linux'
Linux_laptop_data.sort_values(by='Year')
```

```python
laptop_data = pd.concat([windows_laptop_data, MacOS_laptop_data, Linux_laptop_data], ignore_index=True)
laptop_data.drop_duplicates('Laptop_name', inplace = True, ignore_index=True)
laptop_data
```

```python
laptop_data['GPU'] = laptop_data['GPU'].str.replace('NVIDIA ', '')
laptop_data['GPU'] = laptop_data['GPU'].str.replace('AMD ', '')
laptop_data['GPU'] = laptop_data['GPU'].str.replace('Intel ', '')
laptop_data['GPU'] = laptop_data['GPU'].str.replace('ATI ', '')
laptop_data['GPU'] = laptop_data['GPU'].str.replace('Laptop GPU', 'Mobile')
laptop_data['GPU'] = laptop_data['GPU'].str.replace(' SLI', '')
laptop_data['GPU'] = laptop_data['GPU'].str.replace('Graphics Media Accelerator', 'GMA')
```

```python
for i in range(len(laptop_data['GPU'])):
    if type(laptop_data['GPU'].loc[i]) != float:
        if '(' in laptop_data['GPU'].loc[i]:
            laptop_data['GPU'].loc.__setitem__(i, laptop_data['GPU'].loc[i][:laptop_data['GPU'].loc[i].index(' (')])
        if '+' in laptop_data['GPU'].loc[i]:
            laptop_data.drop(i, inplace=True)

laptop_data['GPU'] = laptop_data['GPU'].str.lower()
```

```python
gpu_data = pd.read_csv('raw_data/gpu.csv')
gpu_data.dropna(how = 'all', inplace = True)
gpu_data.drop_duplicates('GPU', inplace = True)
gpu_data['GPU'] = gpu_data['GPU'].str.replace(' IGP', '')
gpu_data['GPU'] = gpu_data['GPU'].str.lower()
gpu_data
```

```python
laptop_data = laptop_data.merge(gpu_data, how='left', on='GPU')
laptop_data.dropna(subset = ['GPU memory'], inplace = True)
laptop_data.reset_index(inplace = True, drop = True)
laptop_data
```

```python
cpu_data = pd.read_csv('raw_data/cpu.csv')
cpu_data['CPU'] = cpu_data['CPU'].str.lower()
cpu_data.rename(columns={'Cores' : 'CPU_Cores', 'Clock' : 'CPU_Clock', 'Process' : 'CPU_Process', 'L3 Cache' : 'CPU_L3', 'TDP' : 'CPU_TDP'}, inplace = True)
cpu_data.drop_duplicates('CPU', inplace = True)
cpu_data
```

```python
laptop_data['CPU'] = laptop_data['CPU'].str.replace('AMD ', '')
laptop_data['CPU'] = laptop_data['CPU'].str.replace('Intel ', '')
laptop_data['CPU'] = laptop_data['CPU'].str.lower()
```

```python
laptop_data = laptop_data.merge(cpu_data, how='left', on='CPU')
laptop_data.dropna(subset=['CPU_Cores'], inplace=True)
laptop_data.reset_index(inplace = True, drop = True)
laptop_data
```

```python
lst_weight =[]
for weight in laptop_data['Weight(kg)']:
    if type(weight) != float:
        if 'kg' in weight:
            lst_weight.append(float(weight[:weight.index('kg')-1]))
        elif 'g' in weight:
            lst_weight.append(float(weight[:weight.index('g')-1])/1000)
        else:
            lst_weight.append(float('nan'))
    else:
        lst_weight.append(float('nan'))

laptop_data['Weight(kg)'] = lst_weight
```

```python
laptop_data['Weight(kg)'].loc.__setitem__(3938, 1.32)
laptop_data['Weight(kg)'].loc.__setitem__(6575, 1.63)
laptop_data['Weight(kg)'].loc.__setitem__(6872, 1.9)
laptop_data['Weight(kg)'].loc.__setitem__(7792, 2.18)
laptop_data['Weight(kg)'].describe()
```

```python
gpu_memory = []
gpu_memory_type = []
gpu_bus_width = []

for gpu_mem in laptop_data['GPU memory']:
    gpu_mem = gpu_mem.split(', ')
    if len(gpu_mem) == 3:
        if 'KB' in gpu_mem[0]:
            gpu_memory.append(float(gpu_mem[0].split()[0])/1024/1024)
        elif 'MB' in gpu_mem[0]:
            gpu_memory.append(float(gpu_mem[0].split()[0])/1024)
        else:
            gpu_memory.append(float(gpu_mem[0].split()[0]))
        gpu_memory_type.append(gpu_mem[1])
        gpu_bus_width.append(gpu_mem[2].split()[0])
    else:
        gpu_memory.append(float('nan'))
        gpu_memory_type.append(float('nan'))
        gpu_bus_width.append(float('nan'))
        
laptop_data['GPU_Memory(GB)'] = gpu_memory
laptop_data['GPU_Memory_Type'] = gpu_memory_type
laptop_data['GPU_Bus_Width(bit)'] = gpu_bus_width
laptop_data.drop(columns=['GPU memory'], inplace=True)
```

```python
lst_cpu_core = []
lst_cpu_thread = []

for cpu_core_thread in laptop_data['CPU_Cores']:
    cpu_core_thread = cpu_core_thread.split(' / ')
    lst_cpu_core.append(int(cpu_core_thread[0]))
    if len(cpu_core_thread) == 2:
        lst_cpu_thread.append(int(cpu_core_thread[1]))
    else:
        lst_cpu_thread.append(int(cpu_core_thread[0]))

laptop_data.drop(columns=['CPU_Cores'], inplace=True)
laptop_data['CPU_Cores'] = lst_cpu_core
laptop_data['CPU_Threads'] = lst_cpu_thread
```

```python
lst_default_clock = []
lst_boosted_clock = []

for cpu_clock in laptop_data['CPU_Clock']:
    cpu_clock = cpu_clock.split(' to ')
    if len(cpu_clock) == 2:       
        if 'MHz' in cpu_clock[1]:
            lst_default_clock.append(float(cpu_clock[0])/1000)
            lst_boosted_clock.append(float(cpu_clock[1].split()[0])/1000)
        else:
            lst_default_clock.append(float(cpu_clock[0]))
            lst_boosted_clock.append(float(cpu_clock[1].split()[0]))
    else:
        lst_boosted_clock.append(float('nan'))
        if 'MHz' in cpu_clock[0]:
            lst_default_clock.append(float(cpu_clock[0].split()[0])/1000)
        else:
            lst_default_clock.append(float(cpu_clock[0].split()[0]))

laptop_data['CPU_Default_Clock'] = lst_default_clock
laptop_data['CPU_Boosted_Clock'] = lst_boosted_clock
laptop_data.drop(columns=['CPU_Clock'], inplace=True)
```

```python
lst_cpu_l3 = []
for cpu_l3 in laptop_data['CPU_L3']:
    if type(cpu_l3) != float:
        lst_cpu_l3.append(float(cpu_l3[:cpu_l3.index('MB')]))
    else:
        lst_cpu_l3.append(float('nan'))

laptop_data['CPU_L3(MB)'] = lst_cpu_l3
laptop_data.drop(columns=['CPU_L3'], inplace=True)
```

```python
laptop_data = laptop_data[[
 'Laptop_name',
 'Year',
 'OS',
 'CPU',
 'CPU_Cores',
 'CPU_Threads',
 'CPU_Default_Clock',
 'CPU_Boosted_Clock',
 'CPU_Process',
 'CPU_L3(MB)',
 'CPU_TDP',
 'GPU',
 'GPU_Memory(GB)',
 'GPU_Memory_Type',
 'GPU_Bus_Width(bit)',
 'GPU clock',
 'GPU memory clock',
 'Memory',
 'Storage',
 'Display',
 'Weight(kg)',
 'Price'
]]
laptop_data
```

```python
laptop_data.to_csv('clean_data/clean_data.csv', index=False)
```
