```python
import pandas as pd
```

```python
data = pd.read_csv('clean_data/clean_data_K.csv')
data
```

```python
lst_gpu_clock = []
for gpu_clock in data['GPU clock']:
    if type(gpu_clock) != float:
        lst_gpu_clock.append(int(gpu_clock.split()[0]))
    else:
        lst_gpu_clock.append(float('nan'))

data['GPU clock'] = lst_gpu_clock
```

```python
lst_gpu_TDP = []
for gpu_TDP in data['CPU_TDP']:
    lst_gpu_TDP.append(int(gpu_TDP.split()[0]))

data['CPU_TDP'] = lst_gpu_TDP
```

```python
lst_gpu_mem_clock = []
for gpu_mem_clock in data['GPU memory clock']:
    gpu_mem_clock = gpu_mem_clock.split()
    if gpu_mem_clock[0] == 'System':
        lst_gpu_mem_clock.append(int(0))
    else:
        lst_gpu_mem_clock.append(int(gpu_mem_clock[0]))
    
data['GPU memory clock'] = lst_gpu_mem_clock
```

```python
lst_storage_type = []
for i in range(len(data)):
    if data['hdd'].loc[i] == 0:
        if data['ssd'].loc[i] == 0:
            lst_storage_type.append(float('nan'))
        else:
            lst_storage_type.append('SSD')
    else:
        lst_storage_type.append('HDD')

data['Storage_Type'] = lst_storage_type
data.drop(columns=['hdd', 'ssd'], inplace=True)
```

```python
data.drop(columns=['Memory', 'Storage', 'Display'], inplace=True)
```

```python
data.rename(columns={'bus_mhz' : 'RAM_bus', 
            'ddr' : 'RAM_DDR', 
            'ram_capacity_gb' : 'RAM(GB)', 
            'storage_capacity_gb' : 'Storage(GB)', 
            'inch' : 'Display_Inch', 
            'pixel_height' : 'Display_Height(pixel)', 
            'pixel_width' : 'Display_Width(pixel)',
            'Price' : 'Price(USD)',
            'GPU clock' : 'GPU clock(MHz)',
            'GPU memory clock' : 'GPU memory clock(MHz)',
            'CPU_TDP' : 'CPU_TDP(W)'}, inplace = True)
```

```python
data = data[[
 'Laptop_name',
 'Year',
 'OS',
 'CPU',
 'CPU_Cores',
 'CPU_Threads',
 'CPU_Default_Clock',
 'CPU_Boosted_Clock',
 'CPU_Transistor_Size(nm)',
 'CPU_L3(MB)',
 'CPU_TDP(W)',
 'GPU',
 'GPU_Memory(GB)',
 'GPU_Memory_Type',
 'GPU_Bus_Width(bit)',
 'GPU clock(MHz)',
 'GPU memory clock(MHz)',
 'RAM(GB)',
 'RAM_bus',
 'RAM_DDR',
 'Storage(GB)',
 'Storage_Type',
 'Display_Inch',
 'Display_Height(pixel)',
 'Display_Width(pixel)',
 'Weight(kg)',
 'Price(USD)'
]]
data
```

```python
# 2022: 1.05115
# 2021: 1.18274
data_2022 = data.where(data['Year'] == 2022).dropna(how = 'all')
data_2021 = data.where(data['Year'] == 2021).dropna(how = 'all')
def price_exchange(data, rate):
    lst_price = []
    for price in data['Price(USD)']:
        try:
            price = price.replace(',', '')
            price = price.replace('.', '')
            if ('EUR' in price) or ('Euro' in price):
                lst_price.append(round(float(price.split('\xa0')[0]) * rate))
            else:
                if 'USD' in price.split()[1]:
                    lst_price.append(float(price.split('\xa0')[0]))
                else:
                    lst_price.append(float(price.split('\xa0')[1]))
        except Exception as e:
            lst_price.append(float('nan'))
    return lst_price
data_2022['Price(USD)'] = price_exchange(data_2022, 1.05115)
data_2021['Price(USD)'] = price_exchange(data_2021, 1.18274)

train_data = pd.concat([data_2022,data_2021], ignore_index=True)
train_data['Price(USD)'].loc.__setitem__(1900, 1681)
train_data['Price(USD)'].loc.__setitem__(142, 1681)
train_data['Price(USD)'].loc.__setitem__(1250, 945)

train_data
```

```python
train_data['Storage_Type'] = train_data['Storage_Type'].fillna('SSD')
train_data['Storage(GB)'] = train_data['Storage(GB)'].fillna(512)
train_data['Weight(kg)'] = train_data['Weight(kg)'].fillna(1.84)
train_data['RAM_DDR'] = train_data['RAM_DDR'].fillna(4)
train_data['RAM(GB)'] = train_data['RAM(GB)'].fillna(16)
train_data[['GPU_Memory(GB)', 'GPU_Bus_Width(bit)', 'RAM_bus']] = train_data[['GPU_Memory(GB)',
                                                            'GPU_Bus_Width(bit)', 'RAM_bus']].fillna(0)
```

```python
train_data['CPU_Cores'] = train_data['CPU_Cores'].astype(int)
train_data['CPU_Threads'] = train_data['CPU_Threads'].astype(int)
train_data['CPU_Transistor_Size(nm)'] = train_data['CPU_Transistor_Size(nm)'].astype(int)
train_data['CPU_L3(MB)'] = train_data['CPU_L3(MB)'].astype(int)
train_data['GPU_Memory(GB)'] = train_data['GPU_Memory(GB)'].astype(int)
train_data['GPU_Bus_Width(bit)'] = train_data['GPU_Bus_Width(bit)'].astype(int)
train_data['RAM(GB)'] = train_data['RAM(GB)'].astype(int)
train_data['RAM_bus'] = train_data['RAM_bus'].astype(int)
train_data['RAM_DDR'] = train_data['RAM_DDR'].astype(int)
train_data['Display_Height(pixel)'] = train_data['Display_Height(pixel)'].astype(int)
train_data['Display_Width(pixel)'] = train_data['Display_Width(pixel)'].astype(int)
train_data['GPU memory clock(MHz)'] = train_data['GPU memory clock(MHz)'].astype(int)
train_data['Storage(GB)'] = train_data['Storage(GB)'].astype(int)
train_data['Year'] = train_data['Year'].astype(int)
train_data['CPU_TDP(W)'] = train_data['CPU_TDP(W)'].astype(int)
```
